```python
import matplotlib.pyplot as plt
import numpy as np

# Data
choices = [
    "Question 1",
    "Question 2",
    "Question 3",
    "Question 4"
]

likert_1 = [10, 30, 22, 16]
likert_2 = [15, 17, 22, 14]
likert_3 = [16, 10, 10, 10]
likert_4 = [14, 3, 6, 14]
likert_5 = [6, 1, 1, 7]

# Define color palette similar to the reference image
colors = ["#FFB3BA", "#FFDFBA", "#FFFFBA", "#BAFFC9", "#BAE1FF"]  # Pastel Red, Orange, Yellow, Green, Blue

# Create figure
fig, ax = plt.subplots(figsize=(8, 4))

# Stacked bar chart
x = np.arange(len(choices))
ax.barh(x, likert_1, color=colors[0], label='1 - Very Low')
ax.barh(x, likert_2, left=likert_1, color=colors[1], label='2 - Low')
ax.barh(x, likert_3, left=np.array(likert_1) + np.array(likert_2), color=colors[2], label='3 - Medium')
ax.barh(x, likert_4, left=np.array(likert_1) + np.array(likert_2) + np.array(likert_3), color=colors[3], label='4 - High')
ax.barh(x, likert_5, left=np.array(likert_1) + np.array(likert_2) + np.array(likert_3) + np.array(likert_4), color=colors[4], label='5 - Very High')

# Add numerical labels inside the bars
for i in range(len(choices)):
    left = 0
    for j, val in enumerate([likert_1[i], likert_2[i], likert_3[i], likert_4[i], likert_5[i]]):
        if val > 0:
            ax.text(left + val / 2, i, str(val), va='center', ha='center', fontsize=9, color='white' if j < 2 else 'black')
        left += val

# Labels and title
ax.set_yticks(x)
ax.set_yticklabels(choices, fontsize=10)
ax.set_xlabel("Number of Responses", fontsize=12, fontweight='bold', color='#2C3E50')

# Aesthetic improvements
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#2C3E50')
ax.spines['bottom'].set_color('#2C3E50')

# Legend placement at the top
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=5, frameon=False, fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
questions = ["Question 1", "Question 2", "Question 3", "Question 4"]
scores = [2.85, 1.82, 2.05, 2.7]

# Create figure and axis
fig, ax = plt.subplots(figsize=(4, 4))

# Plot vertical line chart
ax.plot(scores, np.arange(len(questions)), marker='o', linestyle='-', color='#1E3A8A', linewidth=2, markersize=10)

# Add value labels inside circles
for i, score in enumerate(scores):
    ax.text(score, i, f"{score:.2f}", va='center', ha='center', fontsize=10, color='white', bbox=dict(facecolor='#1E3A8A', edgecolor='none', boxstyle='circle,pad=0.3'))

# Formatting
ax.set_yticks(np.arange(len(questions)))
ax.set_yticklabels(questions, fontsize=10, color='#000000')
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels(["1", "2", "3", "4", "5"], fontsize=9)
ax.set_xlim(1, 5)  # Keep in Likert scale range

# Labels
ax.set_xlabel("Likert Scale Score", fontsize=10, color='#000000')

# Remove unnecessary spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Show the plot
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
